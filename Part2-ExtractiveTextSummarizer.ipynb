```python
import pandas as pd
import numpy as np
import nltk
import re
```

```python
df = pd.read_csv('df_part_i.csv')
df
```

```python
ffull_text = df.iloc[0][4]
```

```python
nltk.download('punkt')
from nltk import tokenize
sentences = tokenize.sent_tokenize(ffull_text)
```

```python
# Lowercase the full text
lst=[]
for s in sentences:
    lst.append(s.lower())
lst
```

```python
# Keep text in your string only
new_lst = []
for s in lst:
    new_lst.append(re.sub('[^A-z]',' ',s))
new_lst
```

```python
# Tokenize sentence: turn sentences into individual words
nltk.download('punkt')
from nltk import tokenize
words = [tokenize.word_tokenize(s) for s in new_lst]
words
```

```python
# Get a dictonary containing the word frequencies

word_freq = {}

for sent in words:
    for word in sent:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1
```

```python
# Get the total weight for each sentence

sent_weight = []
for sent in words:
    weight = 0
    for word in sent:
        weight += word_freq[word]
    sent_weight.append(weight)
sent_weight
```

```python
# Get the indices of the top two weights
indice = (-np.array(sent_weight)).argsort()[:2]
empty_str = ""
for i in list(indice):
    empty_str += sentences[i]
empty_str
```

```python
# Repeat Steps for all text
extracted_summary = []
for i in range(0,1500):
    ffull_text = df.iloc[i][4]
    #split into sentences
    nltk.download('punkt')
    from nltk import tokenize
    sentences = tokenize.sent_tokenize(ffull_text)
    
    #lower caps
    lst=[s.lower() for s in sentences]
    
    #remove non-digit char
    new_lst = [re.sub('[^A-z]',' ',s) for s in lst]
    
    #split list of sentences into list of list of words
  
    words = [tokenize.word_tokenize(s) for s in new_lst]
    #words
    
    #word_freq
    word_freq = {}
    for sent in words:
        for word in sent:
            if word in word_freq:
                word_freq[word] += 1
            else:
                word_freq[word] = 1
    
    #get weight of each sentence
    sent_weight = []
    for sent in words:
        weight = 0
        for word in sent:
            weight += word_freq[word]
        sent_weight.append(weight)
   
    # argsort, get top two sentences and join tgt
    indice = (-np.array(sent_weight)).argsort()[:2]
    empty_str = ""
    for i in list(indice):
        empty_str += sentences[i] + " "
    extracted_summary.append(empty_str)

```

```python
# Print a random 'summary' and the corresponding 'extracted_summary'
print(df.iloc[100][3])
print(extracted_summary[100])
```

```python
# Export the DataFrame as CSV
df['extracted_summary'] = extracted_summary
df.to_csv('part2.csv')
```
